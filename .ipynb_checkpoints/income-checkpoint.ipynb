```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#ignoring the warning sets
import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('adult_dataset.csv')
df.info()
```

```python
df.head()
```

```python
#cleaning our data
#rows that have missing values work class which will affect our decision tree
df_1 = df[df.workclass == '?']
df_1
```

```python
#drop the rows with ? marks
df = df[df['workclass'] != '?']
df.head()
```

```python
#check for categorical values
df_categorical = df.select_dtypes(include = ['object'])
df_categorical.apply(lambda x: x == '?', axis = 0).sum()
```

```python
#drop the categorical values with ? for occupation
df = df[df['occupation'] != '?']
df = df[df['native.country'] != '?']
```

```python
df.info()
```

```python
#Now data preparations
#we have both categorical and numerical features as predictive
#we have to convert categorical values into standard format
#Decission tree can proces categorical values
```

```python
#prepare data
from sklearn import preprocessing

df_categorical = df.select_dtypes(include = ['object'])
df_categorical.head()
```

```python
#convert the categorical values into a standard format by labelencoder
le = preprocessing.LabelEncoder()
df_categorical = df_categorical.apply(le.fit_transform)
df_categorical.head()
```

```python
#Now concatinate our categorical values to our original dataset
#1st drop the categorical values from the original data
#concatinate the new categorical value the dataset
df = df.drop(df_categorical.columns, axis=1)
df = pd.concat([df, df_categorical], axis=1)
df.head()
```

```python
df.info()
```

```python
#convert our target value to category again
df['income'] = df['income'].astype('category')
```

```python
#Now lets build the Model
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting the target into target and training
x = df.drop('income', axis=1)
y = df['income']
```

```python
#now split into x and y train and test
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.30)
```

```python
x_train.head()
```

```python
#classifier 
from sklearn.tree import DecisionTreeClassifier
```

```python
#build decission tree with default hyper parameter then fit
dt_default = DecisionTreeClassifier(max_depth=5)
dt_default.fit(x_train, y_train)
```

```python
#test for accuracy
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
#predict our x_test values
y_pred_default = dt_default.predict(x_test)
```

```python
#calculate classification report
print(classification_report(y_test, y_pred_default))
```

```python
print(confusion_matrix(y_test,y_pred_default))
```

```python
print(accuracy_score(y_test,y_pred_default))
```

```python
#now to print the decision tree you need to installexternal
#library graphviz,pydotplus

from IPython.display import Image
from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz
import pydotplus,graphviz
```

```python
features = list(df.columns[1:])
features
```

```python
#for windows
#import os
#os.environ["path"] +=os.pathsep + 'C:/program Files (x86)....where you installed 
#graphviz'
```

```python
#now make max_depth = 3
dot_data = StringIO()
export_graphviz(dt_default,out_file=dot_data,
               feature_names=features,filled=True, rounded=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

Image(graph.create_png())
```

```python
#The default tree we have is quite complex, so we have to simplifying
#by tuning the hyper parameter
#we can use gini index,information gain(entropy),min_sample_split
```

```python
#Tuning with max_depth
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
#decision tree will stop when reaches 5
n_folds = 5
```

```python
parameters = {'max_depth': range(1,40)}
```

```python
#now let instanciate the model
dtree = DecisionTreeClassifier(criterion = 'gini')
tree = GridSearchCV(dtree,parameters,
                   cv=n_folds,
                   scoring='accuracy')
tree.fit(x_train,y_train)
```

```python
#find the scores
scores = tree.cv_results_
#rf = GridSearchCV(scores, parameters, cv=n_folds,
 #  scoring="accuracy",return_train_score=True)
pd.DataFrame(scores).head()
```

```python
#now plot accuracy with max_depth
plt.figure()
plt.plot(scores["param_max_depth"],
        scores["mean_train_score"],
        label="training accuracy")
plt.plot(scores["param_max_depth"],
        scores["mean_test_score"],
        label="test accuracy")

plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
#the model is overfitting max depth too high
```

```python
#Tuning with min_sample_leaf 
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
n_folds = 5
```

```python
parameters = {"min_samples_leaf": range(5, 200, 20)} 
```

```python
#now let instanciate the model
dtree = DecisionTreeClassifier(criterion = 'gini')
tree = GridSearchCV(dtree,parameters,
                   cv=n_folds,
                   scoring='accuracy')
tree.fit(x_train,y_train)
```

```python
#find the scores
scores = tree.cv_results_
#rf = GridSearchCV(scores, parameters, cv=n_folds,
 #  scoring="accuracy",return_train_score=True)
pd.DataFrame(scores).head()
```

```python
#now plot accuracy with max_depth
plt.figure()
plt.plot(scores["param_min_samples_leaf"],
        scores["mean_train_score"],
        label="training accuracy")
plt.plot(scores["param_min_samples_leaf"],
        scores["mean_test_score"],
        label="test accuracy")

plt.xlabel("min_samples_leaf")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
#the model becomes stable and starts to converge
```

```python
#Tuning with min_sample_split
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
n_folds = 5
```

```python
parameters = {"min_samples_split": range(5,200,20)}
```

```python
#now let instanciate the model
dtree = DecisionTreeClassifier(criterion = 'gini')
tree = GridSearchCV(dtree,parameters,
                   cv=n_folds,
                   scoring='accuracy')
tree.fit(x_train,y_train)
```

```python
#find the scores
scores = tree.cv_results_
#rf = GridSearchCV(scores, parameters, cv=n_folds,
 #  scoring="accuracy",return_train_score=True)
pd.DataFrame(scores).head()
```

```python
#now plot accuracy with max_depth
plt.figure()
plt.plot(scores["param_min_samples_split"],
        scores["mean_train_score"],
        label="training accuracy")
plt.plot(scores["param_min_samples_split"],
        scores["mean_test_score"],
        label="test accuracy")

plt.xlabel("min_samples_split")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
#Now we will use GridSearchCV to find the optimal parameters
```

```python
#create param_grid post
param_grid = {
    'max_depth': range(5,150,5),
    'min_samples_leaf': range(50,150,50),
    'min_samples_split': range(50,150,50),
    'criterion':['entropy','gini']
}
```

```python
n_folds = 5
```

```python
#instaciate the grid search model
dtree = DecisionTreeClassifier()
grid_search = GridSearchCV(estimator = dtree, param_grid = param_grid,
                          cv = n_folds, verbose = 1)
```

```python
#fit the grid search to the data using training values
grid_search.fit(x_train, y_train)
```

```python
#now we take the cv results
cv_results = pd.DataFrame(grid_search.cv_results_)
```

```python
cv_results
```

```python
#Now we will be printing the optimum accuracy score and the hyper parameters
print("best accuracy",grid_search.best_score_)
print(grid_search.best_estimator_)
```

```python
# now run the model using the best parameters 
#obtained fromthe grid_search
clf_gini = DecisionTreeClassifier(criterion = 'gini',
                                 max_depth = 10,
                                 min_samples_leaf = 50,
                                 min_samples_split = 50)
clf_gini.fit(x_train,y_train)
```

```python
#now find accuracy score of clf
clf_gini.score(x_test, y_test)
```

```python
#now plot a tree and see if tree is complex or not
dot_data = StringIO()
export_graphviz(clf_gini,out_file=dot_data, feature_names=features,filled=True,rounded=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
#the tree is too complex,we need to reduce the max_depth
```

```python
#change the max_depth to3
clf_gini = DecisionTreeClassifier(criterion = 'gini',
                                 max_depth=3,
                                 min_samples_leaf=50,
                                 min_samples_split=50)
clf_gini.fit(x_train,y_train)
```

```python
#now  print the accuracy first
print(clf_gini.score(x_test,y_test))
```

```python
dot_data = StringIO()
export_graphviz(clf_gini, out_file=dot_data, feature_names=features, filled=True, rounded=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
y_pred = clf_gini.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```
